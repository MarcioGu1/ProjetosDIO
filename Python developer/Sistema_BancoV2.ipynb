```python
import textwrap
def menu ():
    menu =""" \n
    ================ MENU ================
    [ 1 ]\tDepositar
    [ 2 ]\tSacar
    [ 3 ]\tExtrato
    [ 4 ]\tNova conta
    [ 5 ]\tListar contas
    [ 6 ]\tNovo usuário
    [ 7 ]\tSair
    => """
    return input(textwrap.dedent(menu))
def depositar(saldo,extrato,valor):
    if valor < 0 :
         print(" Valor inválido por favor digite um valor positivo")
    else:
        saldo += valor
        extrato.append(f' Depósito:_______________________________R$  {valor:.2f}'.replace('.',','))
    return saldo ,extrato
def sacar(*,valor,limite = 0,saldo,limite_de_saque =0,extrato =list()):
    if valor <= limite :
        if limite_de_saque > 0:
            if valor <= saldo :
                if valor >= 0:
                    saldo -= valor
                    extrato.append(f" Saque   :_______________________________R$ -{valor:.2f}".replace('.',','))
                    limite_de_saque -= 1
                else :
                    print(f" Por favor digite um valor positivo")
            else :
                print(" O valor é maior do que o valor disponível")  
        else:
            print(" Valor excede o limite de saque diário")
    else :
        print(' O valor é maior que o limite')
    return saldo,extrato,limite_de_saque
def exibir_extrato(saldo = 0,extrato= list()):
    print(f"{'EXTRATO'.center(50)}")
    for info in extrato:
        print(f"\t{info} \n")
    print(f"         Saldo   :_______________________________R$  {saldo:.2f}".replace('.',','))
def find_user(cpf,usuarios):
    usuario_encontrado = [usuario for usuario in usuarios if usuario["cpf"]== cpf]
    return usuario_encontrado[0] if usuario_encontrado else None
def criar_conta(agencia = 0,numero_conta = 0,usuarios = list(),contas = list()):
    cpf = str(input("Informe o CPF do usuário : "))
    usuario = find_user(cpf,usuarios)

    if usuario:
        print(f" < < Conta criada com sucesso > >".center(45))
        contas.append({"agencia": agencia,"numero conta":numero_conta,"usuario":usuario})
    else:
        print(f" Usuário não encontrado, por favor se cadastre-se primeiro")
def listar_users(usuarios):
     for conta in usuarios: 
        print(f"""\
        Agência:\t{conta["agencia"]}
        C/C:\t\t{conta["numero conta"]}
        Titular:\t{conta["usuario"]["nome"]}
        """)
        print("- - "*100)
def novo_user(usuarios = list()):
        cpf = input("Digite seu cpf (apenas numeros): ")
        usuario = find_user(cpf,usuarios)

        if usuario :
            print(" Usúario ja existente")
            return
        
        nome = str(input("Digite seu nome : "))
        data_nascimento = str(input("Ano de nascimento"))
        endereco = str(input('Informe o endereço (logradouro, nro - bairro - cidade/sigla estado): ')) 
        usuarios.append({"nome":nome, "Data":data_nascimento,"cpf":cpf ,"Endereço":endereco})
        print(" Usuário cadastrado com sucesso")
```

```python
def main():
    agencia = "0001"
    saldo = 0
    extrato = []
    usuarios = []
    contas = []
    limite = 500
    limite_de_saque = 3
    while True:
        opcao = menu()
        if   opcao == "1":
            while True:
                try :
                    print(" Qual valor deseja depositar : ")
                    valor = float(input("\nR$ "))
                except (ValueError):
                    print(" Valor inválido pro favor digite um valor válido")
                else:
                    saldo,extrato =depositar(saldo,extrato,valor = valor)
                    break
        elif opcao == "2":
            while True:
                try :
                    print(" Qual valor deseja sacar : ")
                    valor = float(input("\nR$ "))
                except (ValueError):
                    print(" Valor inválido pro favor digite um valor válido")
                else:
                    saldo,extrato,limite_de_saque=sacar(valor =valor,limite = limite, saldo = saldo,limite_de_saque= limite_de_saque,extrato =extrato)
                    break
        elif opcao == "3":
            exibir_extrato(saldo =saldo,extrato =extrato)
        elif opcao == "4":
            numero_conta = len(contas)+1
            criar_conta(agencia = agencia,numero_conta = numero_conta,usuarios = usuarios,contas = contas)
        elif opcao == "5":
            listar_users(contas)
        elif opcao == "6":
            novo_user(usuarios = usuarios)
        elif opcao == "7":
            break
        else:
            print("Opção inválida, por favor digite um número de 1 à 7")

```

```python
main()
```
